```julia
using POMDPs
using Plots
```

```julia
mutable struct PedPOMDP
    size::Tuple{Float64, Float64}
end
```

```julia
mutable struct PedEnv
    m::PedPOMDP
    s::Int
end
```

```julia
function Base.show(io::IO, mime::MIME, env::PedEnv)
    p = plot([0.0,env.m.size[1]], [0.0, env.m.size[2]])
    scatter!([env.s], [0.0])
    show(io, mime, p)
end
```

```julia
m = PedPOMDP((10.0, 1.0))
```

```julia
env = PedEnv(m, 1)
```
